```python
import math
import random

import matplotlib.pyplot as plt
import numpy as np
```

1. Use the next cell to copy and paste your randomwalker class definition.

```python
# Updated the class and method names to be more Pythonic
class RandomWalker():
    def __init__(self):
        # Coordinates
        self.x, self.y = 0.0, 0.0
        # Speed
        self.v = 1.0
    
    def step(self):
        # Each step is a movment of `self.v` in the direction of `angle`
        # skip the middleman and just use radians
        angle = 2*math.pi * random.random()
        self.x += self.v * math.cos(angle)
        self.y += self.v * math.sin(angle)
        
    def distance_from_home(self):
        # Euclidian distance
        return math.sqrt(self.x**2 + self.y**2)
```

2. (3 points) In the next cell, create a function called run() that receives a duration (dur) and a number (n). The function should run a randomwalker "n" times, keeping track at all times of the x,y position of each walker, and their distance from home. The function should return three Numpy arrays: xpos, ypos, and dist. xpos should contain the list of the x positions from the n agents over the duration of walking -- therefore the dimension of this array should be "n" rows by "dur" columns. The same is true for ypos, but for the y positions. And the same is true for dist, except for the distance from home of each agent, at each point in time.

```python
def run(dur, n):
    xpos = np.zeros((dur, n))
    ypos = np.zeros((dur, n))
    dist = np.zeros((dur, n))

    walkers = [RandomWalker() for i in range(n)]
    # 0,0 is initial pos for everyone
    for t in range(1, dur):
        for i, w in enumerate(walkers):
            w.step()
            xpos[(t, i)] = w.x
            ypos[(t, i)] = w.y
            dist[(t, i)] = w.distance_from_home()
    return xpos, ypos, dist
```

10. (3 points) Advanced. Change your randomwalker class (copy and paste it in the next cell, and then modify it) so that instead of turning completely at random, the random walker's new orientation is the old orientation with some Normal noise added to it. The random value should be generated using a Normal distribution with a mean of 0.0 and a standard deviation given by the "alcohol" level of the individual, anywhere between 0.0 (which should lead to a straight path), all the way to about 3.14 (which should lead to a pure random walker as before). Hint: np.random.normal(0.0,self.alcohol).

```python
class RandomWalker():
    def __init__(self, alcohol):
        # Coordinates
        self.x, self.y = 0.0, 0.0
        # Speed
        self.v = 1.0
        self.orientation = 0.0
        self.alcohol = alcohol
    
    def step(self):
        # Each step is a movment of `self.v` in the direction of `angle`
        # skip the middleman and just use radians
        self.orientation += np.random.normal(0.0, self.alcohol)
        self.x += self.v * math.cos(self.orientation)
        self.y += self.v * math.sin(self.orientation)
        
    def distance_from_home(self):
        # Euclidian distance
        return math.sqrt(self.x**2 + self.y**2)
```

11. (3 points) Re-define the run function (again copy and paste) so that in addition to receiving the duration and the number of agents/repetitions, it also now receives the degree of alcohol in the agents.

```python
def run(dur, n, alcohol):
    xpos = np.zeros((dur, n))
    ypos = np.zeros((dur, n))
    dist = np.zeros((dur, n))

    walkers = [RandomWalker(alcohol) for i in range(n)]
    # 0,0 is initial pos for everyone
    for t in range(1, dur):
        for i, w in enumerate(walkers):
            w.step()
            xpos[(t, i)] = w.x
            ypos[(t, i)] = w.y
            dist[(t, i)] = w.distance_from_home()
    return xpos, ypos, dist
```

12. (3 points) Now answer the following question: Does the amount of alcohol in the agent's system affect how far it ends up after a fixed amount of time? If so, how does the velocity of traveling affect the average final distance of the agent? To demonstrate your answer, show a plot that has the different amount of alcohol on the x-axis (from 0.0 to 3.14) and the final average distance from home on the y-axis. Use a duration of 100 and an n of 100. Use linspace to create X levels of alcohol experiments.  

```python
dur = 100
n = 100
alc_levels = 20
dists = np.zeros(alc_levels, dtype=np.float64)
alcs = np.linspace(0, 3.14159, alc_levels)
for i, alc in enumerate(alcs):
    _, __, _dists = run(dur, n, alc)
    dists[i] = np.mean(_dists[-1,:])
```

13. Optional: How does the number of agents used in the experiment affect the results? Show visually.

```python
dur = 100

n_levels = 20
min_n = 10
max_n = 1000

alc_levels = 20

dists = np.zeros((alc_levels, n_levels), dtype=np.float64)
alcs = np.linspace(0, 3.14159, alc_levels)

for i, alc in enumerate(alcs):
    for j, n in enumerate(np.linspace(min_n, max_n, n_levels)):
        _, __, _dists = run(dur, int(n), alc)
        dists[i,j] = np.mean(_dists[-1,:])
```

```python
for j in range(dists.shape[1]):
    plt.plot(dists[:,j], c="k", alpha=0.5)

plt.xlabel("Amount of alcohol")
plt.ylabel("Final distance from home")
plt.title(f"Alcohol varying, {min_n}<n<{max_n}, dur={dur}")
plt.show()

# n varying doesnt change much except that the curves get closer and closer
# to one common curve. This kind of demonstrates the law of large numbers
# from probability theory
```
