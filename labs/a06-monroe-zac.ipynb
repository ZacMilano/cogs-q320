## A06: Agent-Based Model: Thomas Schelling's Segregation Model

In 1971 Thomas Schelling published “Dynamic Models of Segregation,” which proposes a simple model of racial segregation. The Schelling model of the world is a grid; each cell represents a house. The houses are occupied by two kinds of “agents,” labeled red and blue, in roughly equal numbers. About 10% of the houses are empty.

At any point in time, an agent might be happy or unhappy, depending on the other agents in its immediate neighborhood. The neighborhood of each house is the set of eight adjacent cells. In one version of the model, agents are happy if they have at least two neighbors like themselves, and unhappy if they have one or zero.

The simulation proceeds by choosing an agent at random and checking to see whether it is happy. If so, then nothing happens; if not, the agent chooses one of the unoccupied cells at random and moves.

Schelling demonstrated his model with coins on a chess =board. We get the privilege to use a computer. 

The class below is one implementation of the basics of this simulation. (You are welcome to use this one, but you are also welcome to change it, or indeed make your own).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class City():

    def __init__(self, size, proportions, threshold):
        self.size = size        
        self.pop = np.random.choice([-1,0,1],p=proportions,size=(size,size))
        self.threshold = threshold
        
    def show(self):
        plt.imshow(self.pop, interpolation="nearest", cmap="bwr")
        plt.xticks([])
        plt.yticks([])
        plt.show()        
        
    def step(self):
        # pick random house
        i,j = self.randomHouse()
        # find a place to move to if it's unhappy
        if self.numberKin(i,j) < self.threshold:
            self.move(i,j)        
        
    def numberKin(self, i, j):
        return np.count_nonzero(self.pop[i-1:i+2,j-1:j+2]==self.pop[i,j])-1

    def move(self, i, j):
        newi,newj = self.randomVacant()
        self.pop[newi][newj] = self.pop[i][j]
        self.pop[i][j] = 0
        
    def randomHouse(self):
        found = False
        while not found:
            i = np.random.randint(1,self.size-1)
            j = np.random.randint(1,self.size-1)
            if self.pop[i][j] != 0:
                found = True
        return i, j

    def randomVacant(self):
        np.where(self.pop==0)
        found = False
        while not found:
            i = np.random.randint(1,self.size-1)
            j = np.random.randint(1,self.size-1)
            if self.pop[i][j] == 0:
                found = True
        return i, j
           
    def neighbors(self, i, j):
        return self.pop[i-1:i+2,j-1:j+2].flatten()
        
    def segindex(self):
        total = 0
        for i in range(self.size):
            for j in range(self.size):
                nbs = sum(self.neighbors(i,j)) - self.pop[i,j]
                total += np.sign(self.pop[i,j]) *  np.sign(nbs) * nbs ** 2
        return total
```

```python
def run(threshold):
    c = City(50,[0.45,0.1,0.45],threshold)
    moves = 100000
    c.show()
    for i in range(moves):
        c.step()
    c.show()
```

2. (3 points) Define a statistic that measures the degree of segregation. Add this as a method to the class. Call the method segindex(). Plot this statistic over time. Plot it for one threshold of your choice.

```python
def runSeg(threshold, moves, showCity=False, finalOnly=False):
    segs = []
    c = City(50,[0.45,0.1,0.45],threshold)
    if showCity:
        c.show()
    for i in range(moves):
        if not finalOnly:
            segs.append(c.segindex())
        c.step()
    if showCity:
        c.show()
    if finalOnly:
        return c.segindex()
    
    plt.title(f"Segregation index over time, threshold = {threshold}")
    plt.ylabel("Segregation index")
    plt.xlabel("Time step")
    plt.plot(segs)
    plt.show()
```

```python
runSeg(5, 2500, showCity=True)
```

I wonder what the city looks like with more than two classes of populated cells.

**How does the number of classes change the final segregation index with a constant threshold?**

I anticipate that the segregation will be lower in cities that have more class types, so that more diverse places are less segregated.
I will have to completely rewrite the `__init__, show, segindex` methods.

```python
from matplotlib import colors

class City2(City):
    # First proportion is now empty
    def __init__(self, size, proportions, threshold):
        self.size = size        
        # Include empty as a class for easier allocation
        self.nClasses = len(proportions)
        self.pop = np.random.choice(np.arange(self.nClasses),p=proportions,size=(size,size))
        self.threshold = threshold
        
    def show(self):
        cmap = [c for c in "wbrgcmyk"]
        plt.imshow(self.pop, interpolation="nearest", cmap=colors.ListedColormap(cmap[:self.nClasses]))
        plt.xticks([])
        plt.yticks([])
        plt.show()
    
    # Much simpler in principle than the 2-class segindex
    def segindex(self):
        total = 0
        for i in range(self.size):
            for j in range(self.size):
                nbs = self.neighbors(i,j)
                ijClass = self.pop[i,j]
                intermediateTotal = -ijClass # would've just subtracted later, skip that step now
                for n in nbs:
                    if n == ijClass:
                        intermediateTotal += 1
                total += intermediateTotal**2
        return total
```

### Getting an idea of what multiple classes look like on a map

```python
# Still count 0/empty as a class so we need 3 classes at beginning; can only do
# 7 non-empty classes based on color map, so 8 classes max in loop
for nClasses in range(3,9):
    proportions = [(1-0.1)/(nClasses-1) if i != 0 else 0.1
                   for i in range(nClasses)]
    c = City2(50, proportions, 5)
    c.show()
```

```python
segregations = []
for nClasses in range(3,9):
    proportions = [(1-0.1)/(nClasses-1) if i != 0 else 0.1
                   for i in range(nClasses)]
    c = City2(50, proportions, 5)
    for i in range(10000):
        c.step()
    segregations.append(c.segindex())

plt.title("Final segregation as a function of total classes")
plt.xlabel("Number of nonempty classes")
plt.ylabel("Final segregation index")
plt.plot(np.arange(2, 8), segregations)
plt.show()
```
