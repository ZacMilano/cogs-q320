# Matplotlib Reading

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
plt.plot(x, np.square(x))
plt.plot(x, np.square(x), "ro")
plt.ylabel("stonk value")
plt.xlabel("days")
plt.title("Stonks")
plt.show()
```

### Keyword strings
This is the example in the tutorial, copied/pasted, but I altered the variable names to make this example be clearer for me.

```python
data = {
    'a': np.arange(50),
    'color': np.random.randint(0, 50, 50),
    'shape': np.random.randn(50),
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['shape'] = np.abs(data['shape']) * 100

plt.scatter('a', 'b', c='color', s='shape', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Multiple plots, categorical plotting

```python
stonks = ["$MSFT", "$AAPL", "$AMZN"]
mkt_cap = [1.75, 2.23, 1.65]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(stonks, mkt_cap)

plt.subplot(132)
plt.pie(mkt_cap, labels=stonks)

plt.subplot(133)
plt.scatter(stonks, mkt_cap, c="g")

plt.suptitle("Expensive stonks")
plt.show()
```

### Text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# interesting that plt.hist() returns a tuple, and returns anything at all
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('IQ Score')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, f'$\mu={mu},\ \sigma={sigma}$')
# useful
plt.axis([40, 160, 0, 0.03])
# useful
plt.grid(True)
plt.show()
```

### Nonlinear/log axes
This was copied and pasted from the tutorial; I'm tweaking individual parts of it to remind myself of how each part changes when altered.

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(figsize=(6,6))

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear y-scale')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log y-scale')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog y-scale')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit y-scale')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
